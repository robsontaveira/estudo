# Operações Matemáticas

```python
# multiplicação
print(2*2)
# exponencial
print(4**2)
print(pow(4,2))
# divisão
print(10/2)
# módulo (resto da divisão)
print(10%3)
# divisão parte inteira
print(10//3)
```

# Algumas funções


```python
# função type(dado), verifica o tipo do dado ou variável
print(type('a'))
print(type(10))
print(type(10.1))
# função round(number, ndigits)
print(round(3.1415,2))
```
